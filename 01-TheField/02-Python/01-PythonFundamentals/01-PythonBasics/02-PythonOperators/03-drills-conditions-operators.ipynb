**11. Write a script that converts euros into dollars.**
- The program will start by asking the user to indicate with a character 'E' or '$' depending on the currency of the amount they are entering.
- Then the program will ask you to enter the amount and display the conversion.

```python
currency = input("Choose your currency : Euro (€) or Dollar ($) : ")

try: 

    amount = float(input("Enter your amount : "))

    if currency == '€' or currency.lower() == 'euro':
        resultat_euro = amount * 0.92
        print(f"{amount} dollars is {resultat_euro} euro")

    elif currency == '$' or currency.lower() == 'dollar':
        resultat_dollar = amount * 1.08   
        print(f"{amount} euro is {resultat_dollar} dollars")

    else:
        print ("Please enter one of the following terms: Euro, €, Dollar or $")

except ValueError as e:
    print(f" Error : {e}, Please enter a number")


```
